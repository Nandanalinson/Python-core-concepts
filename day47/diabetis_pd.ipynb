```python
# median

# 42,44,47,46,nan,45,48 (odd)

# 1. write in ascending order
# 42,44,45,46,47 -----------> middle value
# here , 45 is the median

# 42,43,44,47,45,48 (even)

# 42,43,44,45,47,48 (ascending order)
#  here ------------> 44+45 / 2 = 44.5 is the median
```
