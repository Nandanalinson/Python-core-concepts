```python
df['education'].unique()  #mode
```

```python
df.fillna({'education': df['education'].mode()[0]}, inplace=True)
df
```

```python
df.isna().sum()
```

```python
df['cigsPerDay'].unique()  #mode
```

```python
df['cigsPerDay'] = df['cigsPerDay'].fillna(df['cigsPerDay'].mode())[0]
df
```

```python
df.isna().sum()
```

```python
df['BPMeds'].unique()  #mode
```

```python
df['BPMeds'] = df['BPMeds'].fillna(df['BPMeds'].mode())[0]
df
```

```python
df.isna().sum()
```

```python
df['totChol'].unique()  #mode
```

```python
df['totChol'] = df['totChol'].fillna(df['totChol'].mode())[0]
df
```

```python
df.isna().sum()
```

```python
df['BMI'].unique()  #mean/median
```

```python
df['BMI'] = df['BMI'].fillna(df['BMI'].mean())[0]
df
```

```python
df.isna().sum()
```

```python
df['heartRate'] = df['heartRate'].fillna(df['heartRate'].mode())[0]
df
```
