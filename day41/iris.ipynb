```python
import pandas as pd
```

```python
df = pd.read_csv('Iris.csv')
```

```python
# 1. Find the number of samples in each category of species
df['Species'].value_counts()
```

```python
# 2. Find the category that it have largest petal length
# df.groupby('Species')['PetalLengthCm'].mean().idxmax()

df.groupby('Species')['PetalLengthCm'].max().tail(1)
```

```python
# 3. Drop the column 'id'
df.drop(columns = ['Id'])
```

```python
# 4. Find the largest sepal length in each category of species
# df.loc[df.groupby(by = 'Species')['SepalLengthCm'].idxmax(), ['Species','SepalLengthCm']]

df.groupby('Species')['SepalLengthCm'].max()
```

```python
# 5. Arrange the data frame in the descending order of sepal length
df.sort_values('SepalLengthCm',ascending=False)
```

```python
# 6. Check is there any missing value
df.isna().sum()
```

```python
# 7. Find the highest sepal width in the category setosa
df[df['Species'] == 'Iris-setosa']['SepalWidthCm'].max()
```

```python
# 8. Find lowest petal length in virginica
df[df['Species'] == 'Iris-virginica']['PetalLengthCm'].min()
```

```python
# 9. Display all the observations where petal length and sepal length greater than 2.5
df[(df['PetalLengthCm'] > 2.5) & (df['SepalLengthCm'] > 2.5)]
```

```python
# 10. Check is there any data with equal value of sepal length and petal length
df[df['SepalLengthCm']==df['PetalLengthCm']].count()
```
