```python
# brand = [maruthi,hydayi,....]
# split
# df['brand'] = df['Name'].str.split().str[0]
# or
```

```python
# apply
# lambda
# lambda argument : expression
df['Brand'] = df['Name'].apply(lambda x:x.split()[0])
# created a new column with an exisiting column ----> feature extraction
```

```python
# object -----> string , to calculate mean , mode or median we have to convert it into int(values)
# int,float

# here , mileage is object dtype , we have to convert it into int. 
# in order to do that , we have to remove the extra km/kg unit "26.6 km/kg"
# so we could get 26.6 as an integer
```

```python
df['Mileage'] = df['Mileage'].str.replace(' km/kg', '')
```

```python
df['Mileage'] = df['Mileage'].str.replace(' kmpl', '')
```

```python
df.dtypes
```

```python
# to convert object into int/float , here errors = coerce means 
# it force any string value to get converted to Nan instead of raising an error
df['Mileage'] = pd.to_numeric(df['Mileage'], errors='coerce')
# or
# df['Mileage'] = df['Mileage'].astype(float)
```

```python
df.dtypes
```

```python
df['Engine'] = df['Engine'].str.replace('CC','')
```

```python
df
```

```python
df['Engine'] = pd.to_numeric(df['Engine'], errors = 'coerce')
```

```python
df.dtypes
```
