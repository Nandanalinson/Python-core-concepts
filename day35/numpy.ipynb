```python
# reshape

y = np.array([[1,2,3],[4,5,6],[9,8,7],[0,0,0]])
y
```

```python
# here there are 12 elements
# to reshape we should have the multiples of 12
# that is , row,col wise : 1*12 , 
                        #  12*1
                        #  2*6
                        #  6*2
                        #  3*4
                        #  4*3
# so here we can have combinations like this only where the product is 12.
```

```python
y.reshape([3,4])    # it is possible bcoz 3*4 = 12
```

```python
y.reshape([6,2])  # here 6 * 2 = 12
```

```python
# now , if we have to reshape into 3D , a new element adds.
# its the depth, and altogether it should be 12
# so here i can we combinations like : 1*6*2
                                    #  1*2*6
                                    #  2*1*6
                                    #  2*12*1
                                    #  3*4*1
                                    #  4*3*1
                                    #  2*12*1
```

```python
y.reshape([1,6,2])
```

```python
y.reshape([2,6,1])
```

```python
y.reshape([3,1,4])
```

```python
y.reshape([4,3,1])
```

```python
y.reshape([12,1,1])
```

```python
y.reshape([6,2,1])
```

```python
y.reshape([2,3,2])
```

```python
y.reshape([3,2,2])
```
