```python
import numpy as np
```

```python
x=np.array([1,2,3,4,8])

y = np.array([2,4,6,8,4])

plt.plot(x,y)
```

```python
h = np.array([43,55,46,77,80])
plt.plot(h)
# default y coordinate is set
```

```python
x1 = np.array([23,34,25,36])

y1=np.array([34,26,43,56])

plt.plot(x1,y1,'o')
# +,*,s-->square,o-->circle
```

```python
# line and point

x2 = np.array([23,45,35,26])
y2 = np.array([27,34,66,28])
plt.plot(x1,y1,marker ='o')
```

```python
x3 = np.array([23,34,45,56])

plt.plot(x3,'g:o')#--->g --green :----dotted line o---circle
plt.plot(x3,'b--o')#b--->blue -- dashed line
plt.plot(x3,'k--s')#k--> black s-->square
```

```python
y = np.array([23,29,28,56])

plt.plot(y,'y:s')#y-->yellow
```

```python
plt.plot(y,'c--s')#c---cyan
```

```python
# marker properties

# ms---> mark size
# mec--->marker edge colour
# mfc--->marker face colour
```

```python
plt.plot(y,marker = 's',ms = 12,mec='b',mfc='g')
```

```python
# line properties

# linestyle ---> --,:
# linewidth
# color
```

```python
plt.plot(y,linestyle = 'dotted',linewidth=2,color='g')
```

```python
# plot line with marker and line properties

d = np.array([45,23,38,78])

plt.plot(d,marker='*',ms = 10,mec='g',mfc='y',linestyle='dotted',linewidth=4,color='y')
```

```python
# two lines in one graph
plt.plot(d,y,marker='+',ms = 10,mec='g',mfc='y',linestyle='dotted',linewidth=4,color='y')
plt.plot(x1,y1,marker = 's',ms = 12,mec='b',mfc='g',linestyle = 'dashed',linewidth=2,color='g')
```

```python
# xlabel
# ylabel

weight=[23,56,65,45]
height=[145,150,156,154]

plt.plot(weight,height)
plt.xlabel('weight',size=12,color='g')
plt.ylabel('height',size= 12,color='r')
plt.title('height vs weight',size=23,color='orange')
plt.grid()
```

```python
# plot graph of study hrs v/s marks

studyhr =[1,3,5,2,6]
marks =[56,69,78,58,90]
plt.plot(studyhr,marks,marker='o',ms=7,linestyle='dotted',color='y')
plt.xlabel('study hour',color='b')
plt.ylabel('marks',color='b')
plt.title('study hours v/s marks',color='r')
plt.grid()
```

```python
# plot graph distance vs time
distance =[1,3,5,9,6]
time =[30,60,70,85,90]
plt.plot(distance,time,marker='*',ms=10,linestyle='dotted',color='y')
plt.xlabel('distance',color='b')
plt.ylabel('time',color='b')
plt.title('distance v/s time',color='k',size =23)
plt.grid()
```

```python
# age vs height

age =[10,37,54,89,60]
height =[130,160,150,185,145]
plt.plot(age,height,marker='*',ms=10,linestyle='dotted',color='k')
plt.xlabel('age',color='b')
plt.ylabel('height',color='b')
plt.title('age vs height',color='orange',size =20)
plt.grid()
```
